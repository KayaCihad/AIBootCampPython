```python
y=df["Sales"]
x=df.drop("Sales",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,train_size=0.80)
```

```python
lr=LinearRegression()
model=lr.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
