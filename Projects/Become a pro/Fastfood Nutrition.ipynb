```python
df["fiber"]=df["fiber"].fillna(df["fiber"].mean())
```

```python
df["protein"]=df["protein"].fillna(df["protein"].mean())
```

```python
df["vit_a"]=df["vit_a"].fillna(df["vit_a"].mean())
```

```python
df["vit_c"]=df["vit_c"].fillna(df["vit_c"].mean())
```

```python
df["calcium"]=df["calcium"].fillna(df["calcium"].mean())
```

```python
df.info()
```

```python
df["sat_fat"]=df["sat_fat"].astype(int)
df["trans_fat"]=df["trans_fat"].astype(int)
df["fiber"]=df["fiber"].astype(int)
df["protein"]=df["protein"].astype(int)
df["vit_a"]=df["vit_a"].astype(int)
df["vit_c"]=df["vit_c"].astype(int)
df["calcium"]=df["calcium"].astype(int)
```

```python
x=df.drop('calories',axis=1)
y=df['calories']
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.sample(10)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g=GaussianNB()
b=BernoulliNB()
```

```python
g.fit(x,y)
b.fit(x,y)
```

```python
tahmin1=g.predict(x)
tahmin2=b.predict(x)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(tahmin1,y)
```

```python
confusion_matrix(tahmin1,y)
```

```python
accuracy_score(tahmin2,y)
```

```python
confusion_matrix(tahmin2,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier()
```

```python
k.fit(x,y)
```

```python
tahmin3=k.predict(x)
```

```python
accuracy_score(tahmin3,y)
```

```python
from sklearn.linear_model import LogisticRegression
l=LogisticRegression()
```

```python
l.fit(x,y)
```

```python
tahmin4=l.predict(x)
```

```python
confusion_matrix(tahmin4,y)
```
